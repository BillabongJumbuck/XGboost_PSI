```python
import pandas as pd
import glob
import os
```

```python
# 读取所有 CSV 文件
csv_files = glob.glob("../data/psi_data_*.csv")
print(f"找到 {len(csv_files)} 个 CSV 文件:")
for f in csv_files:
    print(f"  - {os.path.basename(f)}")

# 合并所有文件
dfs = []
for f in csv_files:
    df_temp = pd.read_csv(f)
    df_temp['source_file'] = os.path.basename(f)
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
print(f"\n总共 {len(df)} 行数据")
```

```python
# 数据预览
print("=== 数据预览 ===")
print(df.head())

print("\n=== 统计信息 ===")
print(df.describe())

print("\n=== 缺失值统计 ===")
print(df.isna().sum())

print("\n=== 每个文件的行数 ===")
print(df.groupby('source_file').size())
```
